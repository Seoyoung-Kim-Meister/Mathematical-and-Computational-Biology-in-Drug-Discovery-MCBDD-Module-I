```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python


# Define the function
def f(x, y):
    return (x * y) / (x * y + (100 - x))

# Create grid of x and y values
# x is the prevalence of disease
# y is the specificity
# The sensitivity is constant = 99%
x = np.linspace(1, 100, 100)
y = np.linspace(1, 100, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create the plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title(r'Visualization of $f(x, y) = \frac{x \cdot y}{x \cdot y + (100 - x)}$')

# Add a color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.savefig('function_plot.png')
```
